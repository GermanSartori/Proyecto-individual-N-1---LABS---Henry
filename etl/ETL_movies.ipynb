```python
import pandas as pd
import ast
```

```python
df_movies = pd.read_parquet(r"E:\Repositorios y bases de datos\Henry DS\Proyecto-individual-N-1---LABS---Henry\datos_procesados\movies_dataset.parquet")
```

```python
df_movies
```

## Desanidado de los datos

```python
df_movies['genres_desanidadas'] = df_movies['genres'].apply(lambda x: ast.literal_eval(x) if isinstance(x, str) else x)
df_movies['belongs_to_collection_desanidada'] = df_movies['belongs_to_collection'].apply(lambda x: ast.literal_eval(x) if isinstance(x, str) else x)
df_movies['production_companies_desanidadas'] = df_movies['production_companies'].apply(lambda x: ast.literal_eval(x) if isinstance(x, str) else x)
df_movies['spoken_languages_desanidadas'] = df_movies['spoken_languages'].apply(lambda x: ast.literal_eval(x) if isinstance(x, str) else x)
df_movies['production_countries_desanidadas'] = df_movies['production_countries'].apply(lambda x: ast.literal_eval(x) if isinstance(x, str) else x)
```

```python
df_movies['genres_desanidadas']
```

```python
# Verificación del tipo de datos antes de la desanidación
print("Tipo de datos antes de la desanidación:")
print(type(df_movies['genres'].iloc[0]))

# Desanidación
if isinstance(df_movies['genres'].iloc[0], str):
    df_movies['genres'] = df_movies['genres'].apply(lambda x: ast.literal_eval(x))

print("\nDespués de la desanidación:")
print(df_movies['genres'].iloc[0])
```

## Eliminación de datos irrelevantes

```python
df_movies
```

```python
df_movies.drop(columns=['adult', 'genres', 'belongs_to_collection', 'homepage', 'imdb_id', 'original_language', 
'overview', 'popularity', 'poster_path', 'production_companies', 
'production_countries', 'runtime', 'spoken_languages', 'status', 'belongs_to_collection_desanidada', 'production_companies_desanidadas', 'spoken_languages_desanidadas', 'production_countries_desanidadas', 
'tagline', 'video'], inplace=True)
```

```python
df_movies_reducido = df_movies
```

```python
df_movies_reducido.dropna(how='all')

df_movies_reducido
```

Limpiando columnas desanidadas

```python
df_movies_limpio = pd.DataFrame(df_movies_reducido)

# Creación de una nueva columna 'genre' con el nombre del primer género de cada película
df_movies_limpio['genre'] = df_movies_limpio['genres_desanidadas'].apply(lambda x: x[0]['name'] if len(x) > 0 else None)

# Eliminación de columnas innecesarias
df_movies_limpio.drop(['genres_desanidadas'], axis=1, inplace=True)

# Muestra del DataFrame resultante
df_movies_limpio
```

## Renombrando y Casteando datos

```python
df_movies_limpio['budget'] = pd.to_numeric(df_movies_limpio['budget'], errors='coerce')
```

```python
df_movies_limpio['release_date'] = pd.to_datetime(df_movies_limpio['release_date'], errors='coerce')
```

```python
print(df_movies_limpio.dtypes)
```

```python
# Valores nulos por columna
print(df_movies_limpio.isnull().sum())
```

```python
# Release_date como datetime
df_movies_limpio['release_date'] = pd.to_datetime(df_movies_limpio['release_date'], errors='coerce')

# Conversión a fecha sin hora
df_movies_limpio['release_date'] = df_movies_limpio['release_date'].dt.date

# Muestra de las primeras filas para verificar
print(df_movies_limpio.head())
```

```python
columnas_null = ['budget', 'genre', 'revenue', 'title', 'release_date', 'vote_average', 'vote_count']

for column in columnas_null:
    df_movies_limpio[column].fillna('Not available', inplace=True)

# Verificación de llenado de los valores nulos
print(df_movies_limpio.isnull().sum())
```

```python
df_movies_limpio
```

```python
# Renombrar la columna 'id' en df_movies a 'movie_id'
df_movies_limpio.rename(columns={'id': 'movie_id'}, inplace=True)
```

```python
df_movies_limpio
```

```python
df_movies_limpio['budget'] = pd.to_numeric(df_movies_limpio['budget'], errors='coerce').fillna(0)

# Creación de la columna 'return' con el cálculo
df_movies_limpio['return'] = df_movies_limpio.apply(
    lambda row: row['revenue'] / row['budget'] if row['budget'] > 0 else 0, axis=1
)

print(df_movies_limpio[['title', 'budget', 'revenue', 'return']].head())
```
