```python
import numpy as np

# Reemplazar los valores no numéricos con NaN
df_movies['movie_id'] = df_movies['movie_id'].replace(['1997-08-20', '2012-09-29', '2014-01-01'], np.nan)

# Convertir movie_id a int64 después de reemplazar con NaN
try:
    df_movies['movie_id'] = df_movies['movie_id'].astype('float').astype('Int64')  # Convertir a tipo Int64
    print("Conversión a int64 exitosa.")
    print(df_movies.dtypes)  # Verificar los tipos de datos después de la conversión
except ValueError as e:
    print(f"Error convirtiendo a int64: {e}")
```

```python
# Merge con cast_df usando 'movie_id' como clave
df_movies = df_movies.merge(df_cast_reducido, on='movie_id', how='left', suffixes=('', '_cast'))

# Merge con crew_df usando 'movie_id' como clave
df_movies = df_movies.merge(df_crew_reducido, on='movie_id', how='left', suffixes=('', '_crew'))

# Verificar las primeras filas del DataFrame resultante
print(df_movies.head())
```
